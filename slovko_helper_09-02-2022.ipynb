### 1. Завантажуємо помічника у Jupiter

```python
import pandas as pd
```

```python
slovko = pd.read_csv('word.csv', encoding='Windows-1251')
slovko.head()
```

```python
print(f'Маємо {slovko.shape[0]} варіантів.')
```

### 2. Розбиваємо слово на літери

```python
slovko['one'] = slovko['word'].astype(str).str[0]
slovko['two'] = slovko['word'].astype(str).str[1]
slovko['three'] = slovko['word'].astype(str).str[2]
slovko['four'] = slovko['word'].astype(str).str[3]
slovko['five'] = slovko['word'].astype(str).str[4]
slovko.head()
```

### 3. Просимо допомогу після першої спроби

```python
first = slovko.query("(one == 'у' | three == 'у' | four == 'у' | five == 'у') & (one == 'о' | two == 'о' | three == 'о' | four == 'о')")
first.head()
```

```python
print(f'Маємо {first.shape[0]} варіантів.')
```

### 4. Просимо допомогу після другої спроби.

```python
second = first.query("(two == 'о') & (one == 'к' | three == 'к' | five == 'к') & (one == 'у' | three == 'у' | four == 'у' | five == 'у')")
second.head(n=100)
```

```python
print(f'Маємо {second.shape[0]} варіантів.')
```

### 5. Просимо допомогу після третьої спроби.

```python
third = second.query("(two == 'о') & (four == 'у') & (one == 'к' | three == 'к' | five == 'к')")
third.head(n=100)
```

```python
print(f'Маємо {third.shape[0]} варіантів.')
```

### 6. Просимо допомоги після четвертої спроби.

```python
fourth = third.query("(two == 'о') & (four == 'у') & (three == 'к' | five == 'к')")
fourth.head(n=100)
```

```python
print(f'Маємо {fourth.shape[0]} варіантів.')
```

### 7. Просимо допомогти після п'ятої спроби.

```python
fifth = fourth.query("(two == 'о') & (four == 'у') & (three == 'к' | five == 'к') & (one == 'п')")
fifth.head(n=100)
```

```python
print(f'Маємо {fifth.shape[0]} варіантів.')
```
