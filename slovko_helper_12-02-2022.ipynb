### 2. Розбиваємо слово на літери і додаємо ваги кожному

```python
slovko['one'] = slovko['word'].astype(str).str[0]
slovko['two'] = slovko['word'].astype(str).str[1]
slovko['three'] = slovko['word'].astype(str).str[2]
slovko['four'] = slovko['word'].astype(str).str[3]
slovko['five'] = slovko['word'].astype(str).str[4]
slovko.head()
```

```python
def mul_round(df):
    return df.mul(100).round(2)
```

```python
def slovko_prob(df):

    one_df = df['one'].value_counts(normalize=True).to_frame()
    one_df = one_df.rename(columns={'one': 'one_norm'})
    one_df['one'] = one_df.index

    two_df = df['two'].value_counts(normalize=True).to_frame()
    two_df = two_df.rename(columns={'two': 'two_norm'})
    two_df['two'] = two_df.index

    three_df = df['three'].value_counts(normalize=True).to_frame()
    three_df = three_df.rename(columns={'three': 'three_norm'})
    three_df['three'] = three_df.index

    four_df = df['four'].value_counts(normalize=True).to_frame()
    four_df = four_df.rename(columns={'four': 'four_norm'})
    four_df['four'] = four_df.index

    five_df = df['five'].value_counts(normalize=True).to_frame()
    five_df = five_df.rename(columns={'five': 'five_norm'})
    five_df['five'] = five_df.index
    
    df = df.merge(one_df, how='left', on='one')
    df = df.merge(two_df, how='left', on='two')
    df = df.merge(three_df, how='left', on='three')
    df = df.merge(four_df, how='left', on='four')
    df = df.merge(five_df, how='left', on='five')
    
    df['one_norm'] = mul_round(df['one_norm'])
    df['two_norm'] = mul_round(df['two_norm'])
    df['three_norm'] = mul_round(df['three_norm'])
    df['four_norm'] = mul_round(df['four_norm'])
    df['five_norm'] = mul_round(df['five_norm'])
    
    df['sum'] = (df['one_norm'] + df['two_norm'] + df['three_norm'] + df['four_norm'] + df['five_norm']) / 5
    df['prob'] = (df['sum'] / df['sum'].sum()).mul(100).round(3)
    df = df.sort_values(by='sum', ascending=False)

    return df
```

```python
slovko_prob(slovko)
```

### 3. Просимо допомогу після першої спроби

```python
first = slovko.query("(one == 'к')")
first = first.query("(one != 'а' & two != 'а' & three != 'о' & four != 'а' & five != 'а')")
first = first.query("(one != 'р' & two != 'р' & three != 'р' & four != 'р' & five != 'р')")
first = first.query("(one != 'н' & two != 'н' & three != 'н' & four != 'н' & five != 'н')")
first = first.query("(one != 'а' & two != 'а' & three != 'о' & four != 'а' & five != 'а')")
```

```python
first = first[['word', 'one', 'two', 'three', 'four', 'five']]
first.head()
```

```python
slovko_prob(first)
```
